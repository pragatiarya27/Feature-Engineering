#4. How does loss value help in determining whether the model is good or not?

- Low loss value → model predictions close to actual values → good model

- High loss value → poor predictions
The loss tracks the training progress and tells the optimizer which direction to update parameters.

#5. What are continuous and categorical variables?

- Continuous: Numerical values with infinite possible values (e.g., height, salary).

- Categorical: Non-numerical categories or labels (e.g., gender, color, city).

#6. How do we handle categorical variables? What are the common technique?
- Categorical variables contain non-numerical values (like gender, color, city).
Since machine learning models work with numbers, we must convert categories into numeric form.
- This process is called encoding.
- Common encoding techniques:

- Label Encoding

- One-Hot Encoding

- Ordinal Encoding

- Binary Encoding

- Target Encoding (for advanced ML)

#15. What is causation?  Explain difference between correlation and causation with an example.

- Causation: One variable directly affects another.

- Difference:

Correlation → variables move together

Causation → one variable produces change in the other

Example:
Ice cream sales ↑ and drowning cases ↑
They are correlated (summer) but ice cream does not cause drowning.
This is correlation, not causation.

#21. What is feature scaling? How does it help in Machine Learning?

- Feature scaling transforms variables to the same scale.

- Helps by:

- Speeding up training

- Improving model performance

- Stabilizing gradient descent

- Avoiding dominance of large-value features

#22. How do we perform scaling in Python?
